```python
import yfinance as yf
import pandas as pd
```

```python
nvda = yf.Ticker("NVDA")
```

```python
nvda = nvda.history(period="max", interval="1wk")
```

```python
nvda
```

```python
nvda['Open %change'] = nvda['Open'] / nvda['Open'].shift(1)
nvda['Close %change'] = nvda['Close'] / nvda['Close'].shift(1)
nvda['Low %change'] = nvda['Low'] / nvda['Low'].shift(1)
nvda['High %change'] = nvda['High'] / nvda['High'].shift(1)
nvda['Volume %change'] = nvda['Volume'] / nvda['Volume'].shift(1)
nvda = nvda.drop(nvda.index[0])
nvda
```

```python
nvda.index
```

```python
nvda.plot.line(y="Close", use_index=True)
```

```python
del nvda["Dividends"]
del nvda["Stock Splits"]
```

```python
nvda["NextYear"] = nvda["Close"].shift(-50)
```

```python
nvda["Target"] = nvda["NextYear"]
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["Close", "Volume", "Open", "High", "Low", "Open %change", "Close %change", "Low %change", "High %change", "Volume %change"]
```

```python
train = nvda.iloc[:-50]
test = nvda
```

```python
reg.fit(train[predictors],train["Target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
combined = pd.concat([test["Target"], pd.Series(predictions, index=test.index)], axis=1)
```

```python
combined.columns = ["actual", "predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
myResults = combined.iloc[-60:]
```
